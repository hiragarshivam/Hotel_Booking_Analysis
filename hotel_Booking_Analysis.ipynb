  CLEANING THE DATA:

```python
df.drop(columns=["company", "agent"],inplace = True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()           #This is applicable to numeric values.
```

```python
df=df[df["children"]<5]              #Removing outliers from the given Data set.
df=df[df["babies"]<5]
df=df[df["adr"]<5000]
df=df[df["adr"]>48]
```

```python
plt.figure(figsize=(5,5))                  #Visualization of above counts.

x=["Not cancelled","Cancelled"]
y=df["is_canceled"].value_counts()
 
plt.bar(x,y,color="#2ca876")
plt.title("Reservation status count",size=18,loc="left")
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(9,5))               #Visualization of cancellation in both types of hotel.

sns.countplot(x="hotel",hue="is_canceled",data=df,palette="viridis")

plt.title("Reservation status in different Hotels",size=18,loc="left")
plt.xlabel("Hotel")
plt.ylabel("Number of Reservations")
plt.legend(["Not cancelled","Cancelled"])
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']             #Filtering of hotel to resort hotel.
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']                 #Filtering of hotel to city hotel.
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Grouping of Data as per the date of reservation and found the mean adr of that day.

resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize=(20,8))

plt.title('Average daily rates in city and resort hotels',size=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel',color="#2ca876")
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel',color="#406484")
plt.legend(fontsize="xx-large")
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month   #Creating a column for months where dt.month will take the month from
                                                     #datetime format of given column.

plt.figure(figsize=(16,8))
sns.countplot(x='month',hue='is_canceled',data=df,palette='viridis')

plt.title('Reservation status in per month',size=18)
plt.xlabel('Month',size=20)
plt.ylabel('Number of Reservations')
plt.legend(['Not cancelled','Cancelled'],fontsize="xx-large")
plt.show()
```

```python
# The y-axis represents the total cost of canceled reservations ("adr") for each of those months. The code first filters the data to consider only canceled reservations, then groups the data by month, calculates the total cost for each month:
plt.figure(figsize=(15, 8))
plt.title('ADR per month', fontsize=20)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index(),palette="viridis")
plt.show()
```

```python
#Visualizing Top 10 Countries with Canceled Reservations:

cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[0:10]
plt.figure(figsize=(8,8))
plt.title('Top 10- Countries with Reservation cancelled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index,shadow=True)
plt.show()
```

CONCLUSION: PRICE IS EFFECTING THE CANCELLATION RATE:

```python
#Analyzing Average Daily Rate (ADR) Trends:
cancelled_df_adr = cancelled_data.groupby("reservation_status_date")[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df["is_canceled"] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby("reservation_status_date")[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values("reservation_status_date", inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr["reservation_status_date"], not_cancelled_df_adr['adr'], label='not cancelled',color="#2ca876")
plt.plot(cancelled_df_adr["reservation_status_date"], cancelled_df_adr['adr'], label='cancelled',color="#406484")
plt.legend(fontsize="xx-large")
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')& (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')& (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```
